# Python built in math functions and math module

Python has a set of built-in math functions, and a math module, for extensive mathematical tasks.

```python
# Example 1, min(). max(), abs(), pow(), sum(), divmod()
minV = min([3, 5, 7]); maxV = max([3, 5, 7])
print("Maximum value is: ", minV, "Minimum value is: ",maxV)

absV = abs(-3.77)
print("Absolute value is: ", absV)
powV = pow(2, 4)
print("The power value is: ",powV)
sumV = sum([1, 2, 3])
print("The sum is: ", sumV)
result = divmod(10, 3) # Returns the quotient and the remainder
print("The output of divmod: ",result)
```

```python
# Example 2, round()
print(round(10.1564, 3)) 
print(round(10.1554, 6)) 
print(round(1723.155411, 0))
```

```python
# Example 4, sqrt(), ceil(), pow(), remainder(), trunc()
import math

print("The square root is: ", math.sqrt(64))
print("Ceiling value is: ", math.ceil(1.4) , "floor value is: ", math.floor(1.4))
print("Power", math.pow(2, 3))                
print("Remainder is: ", math.remainder(10, 3)) 

print("Truncated integer value is: ", math.trunc(10.335)) 
# Returns the truncated integer parts of a number
```

```python
# Example 5, comb(), perm(), copysign(), dist(), exp(), factorial(), fmod(), frexp()
import math

print("Total number of nCk: ", math.comb(10,5))     # comb(n,k)
print("Total nPk value: ", math.perm(10, 5))        # perm (n,k)
print("The value is: ",math.copysign(-2, 5)) 

print("Euclidean distance is: ",math.dist([2, 3], [4, 6])) # the Euclidean distance 
print("e^x: ", math.exp(3))                       # E to the power of x, exp(x)
print ("Factorial value is: ", math.factorial(6)) # the factorial of a number n, fact(n)
print("Remainder value is: ", math.fmod(100, 7))   # Returns the remainder of x/y, fmod(x,y)
print("The mantissa and the exponent",math.frexp(2.0)) 
# Returns the mantissa and the exponent, of a number
```

```python
# Example 6, fsum(), prod(), gcd(), isfinite(), isinf(), isnan(), isqrt()
import math

print("Sum is: ", math.fsum([1, 4, 5])) # Returns the sum 
print("Mul is: ", math.prod([1, 4, 5])) # Returns the product
print("The GCD value is: ", math.gcd(12, 16)) # Returns the greatest common divisor 

print("isfinite: ", math.isfinite(10/3)) # Checks whether a number is finite or not
print("isinf: ", math.isinf(5/3))  # Checks whether a number is infinite or not
print("isnan: ", math.isnan(10))   # Checks whether a value is NaN (not a number) or not
print("isqrt: ", math.isqrt(10))   # Rounds a square root number downwards to the nearest integer
```

```python
# Example 7, log
import math

print("Log base e:", math.log(10))                 # Returns the natural logarithm of a number
print("Log base 10", math.log10(10))               # Returns the base-10 logarithm of x
print("Log base e: ", math.log1p(9))               # Returns the natural logarithm of 1+x
print("Log base 2: ", math.log2(10))                # Returns the base-2 logarithm of x
print("Log base x:", math.log(8,3))                 # Return log2(8), log x base b, log(x,b)
```

```python
# Exampl 8, constant and their values
import math

print(math.e)     # Returns Euler's number (2.7182...)
print(math.inf)   # Returns a floating-point positive infinity
print(math.nan)   #Returns a floating-point NaN (Not a Number) value
print(math.pi)    #Returns PI (3.1415...)
print(math.tau)   #Returns tau (6.2831...)

r = 10
print(math.pi * math.pow(r,2))
```

```python
# Example 9, sin(), cos(), tan(), radians(), degrees()
import math

print(math.sin(1.5707963267948966))          #  sine of a number, sin(radianValue)
print(math.cos(1.5707963267948966))          #  cosine of a number, cos(radianValue)
print(math.tan(0.785398))                    #  tangent of a number, tan(radianValue)

print(math.radians(90))          #  Converts a degree value into radians
print(math.degrees(1.5708))      #  Converts an angle from radians to degrees
print(math.radians(45)) 

#math.sinh()         # Returns the hyperbolic sine of a number
#math.cosh()         # Returns the hyperbolic cosine of a number
#math.tanh()         # Returns the hyperbolic tangent of a number
```

# Other useful functions
* math.erf()	Returns the error function of a number
* math.erfc()	Returns the complementary error function of a number
* math.gamma()	Returns the gamma function at x
* math.hypot()	Returns the Euclidean norm
* math.isclose()	Checks whether two values are close to each other, or not
* math.ldexp()	Returns the inverse of math.frexp() which is x * (2**i) of the given numbers x and i
* math.lgamma()	Returns the log gamma value of x

# Python Built in Functions

```python
# Example 10, all(), any(), sorted()
print(all([1, 2, 3, range(2), 'A']))  # Returns True if all items in an iterable object are true

print(any([1, 23, 'A']))  # Returns True if any item in an iterable object is true

print(sorted([7,4,9,2]))  # Returns a sorted list 

print(sorted([7,4,9,2], reverse = True)) # Returns a sorted lis
```

```python
# Example 11, filter()
myList = [0, 1, 2, 3, 8]

result = filter(lambda x: x % 2 == 0, myList)

print("1.",list(result))

def oddCheck(val):
    if val%2 != 0:
        return True
    
result = filter(oddCheck, myList)
print("2.",list(result))
```

```python
# Example 12, enumerate() 
myList = ["Off the Wall", "Thriller", "Dangerous"]
 
# printing the tuples in object directly
for i in enumerate(myList):
    print (i)
```

```python
# Example 13, zip()    
M = ("Ram", "Krishna", "Mahadev")
F = ("Sita", "Radha", "Parvathi")

om = zip(M,F)
for i in om:
    print (i)
```

```python
# Example 14, iter(), reversed()
x = iter(["Georgina", "Hilton", "Kim"])
print(next(x))
print(next(x))

y = reversed(["__Selena", "__Sofia", "__Kendall"])
print(next(y))
print(next(y))
```
