##  1.1 If statement
* Statements under if will be executed if the condition is true.
> if condition:
>> statement(s)

```python
# Example 1 if condition
number = 5

print(number >=0)

if number >=0:                          # IndentationError: expected an indented block
    print("Number is positive.")
    
print("Hi! There.")
```

```python
# Example 2, if condition
number = int(input("Enter number:"))

print(number%2 == 0)

if number%2 == 0:
    print("Number is even number.")
```

```python
# Example 3, if condition
data = input("Enter number:")

if len(data) > 2:
    print("The size of the data is more than 2")
     
if len(data) >= 5 and len(data)<= 7:
    print("The size of the data is in between 5 and 7")
    
if len(data) >=5 or data.isdigit():
    print("The size of the data is more than 5 or a digit")
```

```python
# Example of not
if ~False:
    print("If statement is executed!!!")
print("Not is going on!")
```

## 1.2 if - else
> if condition :
>> statement x 

> else : 
>> statement y

* If the condition is true, statement xwill be executed; otherwise, statement ywill be executed.

```python
# Example 4, if else condition
number = 50
print("Condition status:",number >=0)

if number >=0:
    print("Number is positive.")
    
else:
    print("Number is negative.")
```

```python
# Example 5, if else condition
number = 6

if number%2 == 0:
    print("Number is even number.")
    
else:
    print("Number is odd number.")
    
print("The Academician")
print("if else is going on")
```

```python
# Example 6, if else condition
data = input("Enter number:")

if len(data)> 5 and len(data) < 8:
    print("The size of the data is between 5 and 8")
    
else:
    print("The size of the data is less than 5 or more than 7")
```

## 1.3 if-elif-else

```python
# Example 7, if elif else condition
number = int(input("Enter number:"))

if number > 0:
    print("Number is more than zero.")
    
elif number < 0:
    print("Number is less than zero.")
    
else:
    print("Number is zero.")
    
print("if-elif-else is going on")
```

### 2. Loops
* Execute a group of lines multiple times. 
* for, while, for-else, while-else

# 2.1 for loop
> for x in sequence :
>>  lines to execute

Note: sequence can be (any collection) list, tuple, set, dictionary, or string.
Group of lines will be executed for every data present in the sequence. Instead of the typical newline character, the end = ' ' causes a space to be displayed after each number.

```python
# Example 10, for loop
for i in range(3):
    print(i, end =' ')

print()
for i in range(-2, 5):
    print(i, end =' ')
```

```python
#print()
for i in range(2, 10, 2):
    print(i, end =' ')
    
print()
for i in range(5, 0, -1):
    print(i, end =' ')
```

```python
# Example 11, for loop
myString = "Baseball"
for char in myString :
    print(char, end = ' ') 

print()  
for index in range(len(myString)):
    print(myString[index], "is", index, end = ', ') 
```

```python
# Example 12, for loop
seq  = [3, 5, 7, 9] 
print("Data type:", type(seq))

for data in seq:
    print("The content of seq is: ",data)
    
for data in [13, 25, 37, 49]:
    print(data, end = ' ')
```

```python
# Example 13, for loop in set
seq  = {3, 5, 7, 9} 
print("Data type:", type(seq))

for data in seq:
    print("The content of seq is: ",data)
```

```python
# Example 14, for loop in tuple
seq  = (3, 5, 7, 9) 
print("Data type:", type(seq))

for data in seq:
    print("The content of seq is: ",data)
```

```python
# Example 15 Dictionary iteration
seq  = {1:'A', 2:'B', 3:'C'}
print("Data type:", type(seq))

for k,v in seq.items(): # need both key and value
    print(k, ":", v, end = '      ')
   
print() 
for v in seq.values(): # need only values
    print(v, end = ' ')

print()    
for i in seq: # need only keys
    print(i, end = ' ')
```

```python
# Example 16 Nested Loop
for i in range(1,4):
    for j in range(1,4):
        print("i=", i ," j=", j) 
```

```python
# Example 17 Nested Loop
for i in range(1,4):
    for j in range(i,4):
        print("i=", i ," j=", j) 
```

```python
# For with place holder
for _ in range(3):
    print("Yes", end = ' ')
```

## 2.2 while
To execute a group of line repeatedly until condition false.

Syntax:
> while condition :
>>  group of lines

```python
# Simple while example
i = 0 # Initialization
while i <= 4: # define the condition
    print(i, end = ' ')
    i +=1  # change  
```

```python
# Example 18, while loop
sum = 0; i = 1
while i <= 5:
    print(i, end = ' ')
    sum += i 
    i += 1

print("\nThe sum is: ", sum)
```

```python
# Nested while loop
i = 0; j = 0
while i < 10:
    while j < i:
        print (i, j, end = '   ::')
        j += 1
    i+=1
```

## 3.1 Break
* Inside loops, the break statement interrupts loop execution based on a condition.

```python
# Example 19, break
i = 0
while True:
    print("Loop is executing", i)
    i +=1
    if i == 5:
        break
```

```python
# Example 20, break
import random

k = 0
while True:
    x = random.randint(1, 10)
    print(x, end = ', ')
    
    k +=1
    if x == 5:
        print("Bingoooo!")
        print("'Times'", k-1)
        break
```

```python
# Example 21, break
for i in range(10):
    print(i, end = ' ')
    if i == 3:
        break
```

```python
# Example 21, break
for i in 'abbdbba':
    print(i , end = ' ')
    if i == 'd':                       # What happened if I put if - break before print?
        break
        
        
```

## Continue
* to skip current iteration and continue for the next iterations

```python
# Example 22, continue
i = 0
while i<6:
    i += 1
    if i %2 == 0:
        continue
    print(i, end = ' ')
```

```python
# Example 23, continue
list1 = [1, 7, 9, 5, 4]

for i in range(10):
    if i in list1:
        continue
    print(i, end = ' ')    
```

## For- else, While- else
* else is just loop without break

> for x in sequence :
>>  statements x

>else:
>>  statements y

```python
# Example 24, for-else
list1 = [3, 9, 6, 8]

for i in range(5):
    print("Loop is executing", i)
    if i in list1:
        print("Break is executed!")
        break
else:
    print("Break is not executed!")
    print("Processed all the elements")
```

```python
# While with else 

k = int(input("Enter value for k:"))
i = 0
while i<=10:
    print(i, end = ' ')
    i +=1
    if i == k:
        break
else:
    print("\nWhile loop is executed completely, and value of k is more than 10.")
```

## 3.3 Pass
* The Pass is a null statement. The interpreter does not ignore a "pass" statement, and the statement has no effect. When you don't write the implementation of that part of the program but intend to do it later.

```python
# Example 25, pass
for i in range(5):
    pass

else:
    print("Processed all the elements")
```

```python
# Example 26, pass
number = 1

if number%2 == 0:
    print("Number is even number.")
    
else:
    pass   
```
