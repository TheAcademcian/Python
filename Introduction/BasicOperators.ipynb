# Basic Operators
Operators are symbol that performs some specific operations.

**Operators**
* 1. Arithmetic
* 2. Relational or Comparison
* 3. Logical
* 4. Bitwise
* 5. Assignment
* 6. Special

## 1. Arithmetic Operators
\+ Addition, - Subtraction, * Multiplication, / Division, % Modulo, // Floor Division operator, ** Exponent or power operator

```python
# Example 1, Arithmetic Operators
x, y = 6, 3
#x = 7; y = 3
#x, y = 4.5, 7.8
#x, y = 'ABC', 'XYZ'

print("x:",x, "y",y)
print('x + y:', x + y)
print('x - y:', x - y)
print('x * y:', x * y)
print('x / y:', x / y) # / operator always performs floating point arithmetic, 16 digit
print('x //y:', x // y)
print('x % y:', x % y)
print('x ** y:', x ** y) 
```

## 2. Relational Operators

```python
# Example 2, Relational Operators
x = 10; y = 20
#x = 7.9; y = 3.1
#x = "Ganga"; y = "Yamuna"
#x  =  True; y = False
#x  =  3.5; y = False
#x  =  3.5; y = 'Salsa'
print("x:",x, "y:",y)
print("x > y : ", x > y)
print("x >= y: ", x >= y)
print("x < y : ", x < y)
print("x <= y: ", x <= y)
print("x == y: ", x == y)
print("x != y: ", x != y)
```

```python
# Example 3, Relational Operators
x = 10; y = 20; z = -3

print( x > y < z) # If atleast one comparison returns False then the False is returned.
print( x < y > z)
print( x != y != z)
print( x != y == z)
```

## 3. Logical Operators
* and, or ,not. (False: 0, True: non-zero)

```python
# Example 4, Logical Operators, 
x = 10; y = 15 # 
#x = -10; y = 20 # x  = 'ABC'; y = 'XYX'

print("x:",x, "y:",y) 
print( x<10 and y>0) 
print( x<10 or y>0) 
print( not x<20)

print( (x<10) or (y>0 and x==y) )
```

## 4. Bitwise Operators
* only for int and boolean types.
* &, |, ^, <<, >>, ~

```python
# Example 5, Bitwise Operators
x = 10; y = 2
# x = True; y = False
# x = True; y = True
#x = -10; y = 2
#x = "River"; y = "Ocean"
#x = 10.4; y = 2.5

print("x & y:", x & y) # AND
print("x | y:", x | y) # OR
print("x ^ y:", x ^ y) # Ex-Or
print("x >> y:",x >> y) # Right shift y times
print("x << y:", x << y) # Left shift y times
print("~x :", ~x) # NOT
```

## 5. Assignment Operators
* = to assign a value to a variable

```python
# Example 6, Assignment Operators
x = 77
print("The content of x is: ", x, type(x))
```

```python
# Example 7, Increment and decrement
x = 5
x = x + 1
print("The content of x is: ", x)

x += 1
print("The content of x is: ", x)

x = x - 1
print("The content of x is: ", x)

x -= 1
print("The content of x is: ", x)
```

```python
# Example 8, Multiplication, devision, power 
x = 5
x = x * 2
print("The content of x is: ", x)
x *= 2
print("The content of x is: ", x)
x /= 2
print("The content of x is: ", x)
x **= 2
print("The content of x is: ", x)
x //= 2
print("The content of x is: ", x)
```

```python
# Example 9, assignment
# x = 10
# y = 12

x, y = 10, 12
print("The content of x is: ", x)
print("The content of y is: ", y)

a, b, c, d = 7, 9, 8, -2
print("The variables are: ", a,b,c,d)
```

## Ternary Operator
Syntax:
> varName = value_1 if condition else value_2

```python
# Example 10, Ternary Operator
#x, y = 10, 5
x, y = 3, 9

myVar = x if x > y else y
#myVar = True if x > y else False

print("Bigger value is: ", myVar) 
```

```python
# Example 11, Nested ternary operator
x = int(input("Enter number:"))

print("Number = zero." if x == 0 else "Number > zero." if x > 0 else "Number < zero.")
```

```python
# Example 12 Nested Ternary Operator (Maximum and minimum of three numbers)
x=int(input("Enter number for x:"))
y=int(input("Enter number for y:"))
z=int(input("Enter number for z:"))

minNumber = x if x<y and x<z else y if y<z else z
maxNumber = x if x>y and x>z else y if y>z else z

print("\nMinimum number is:", minNumber) 
print("Maximum number is:", maxNumber) 
```

# 6. Special
* Identity Operators: for address comparison
* Membership operators: to check whether the given element is present in the given collection (list, set, tuple, dictionary, string)

```python
# Example 13 Identity Operators
#x = 10; y = 15
x = 10; y = 10

print("Address of x: ",id(x))
print("Address of y:",id(y))
print(x is y) # True if both the variables pointing to the same object
```

```python
# Example 14 Membership operators
coll = [3, 5, 7, 8]
#coll = {3, 5, 7, 8}
#coll = (3, 5, 7, 8)
#coll = '3578'; element = '3'
element = 3
print(element in coll)
print(element not in coll)

if element in coll:
    print("1.Element is present in the coll!")
else:
    print("1.Element is not present in the coll!")
    
if element not in coll:
    print("2.Element is not present in the coll!")
else:
    print("2.Element is present in the coll!")
```
