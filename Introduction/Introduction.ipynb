# Identifiers
* Any name is referred to as an identifier e.g., class name, function name, module name, variable name
* Rules: a to z, A to Z, 0 - 9, and _ 
* Length limit? 79 Characters 

```python
# Example 2, Rules for identifiers
Boss = 1
Boss1 = 2

boss, BOSS, bOSS, Boss = 4, 5, 6, 7
print(Boss, Boss1, boss, BOSS, bOSS, Boss)

boss_ = 8
_boss = 9                   # private variable
hi_boss = 10
__boss = 11                 # strongly private identifier
__boss__ = 12               # special name, which is also known as magic methods.
print(boss_, _boss, hi_boss, __boss, __boss__)

#123 = 10
#1Boss = 3
#def = 9 # cannot use reserved words as identifiers
#if = 9
#boss@ = 9
#boss$ = 9
```

Note:
1. Identifier starts with \_ symbol (single underscore): private variable
2. Identifier starts with \_\_ symbol (double underscore): strongly private identifier.

3. Identifier starts and ends with \_\_  (double underscore): define as special name, which is also known as magic methods.

# Data Types
* Says type of data present inside a variable.
* Python is a dynamically typed programming language: The type will be assigned to the variable automatically.
*  int, float, complex, bool, str, bytes, bytearray, range, list, tuple, set, frozenset, dict, None


```python
# Example 4, data type
a = 7
print(a, type(a))
a = 7.5
print(a, type(a))
a = 1.2E3               # a = 1.2e3
print(a, type(a))

a = 7 + 6j
print(a, type(a))

a = True
print(a, type(a))

a = 'Himalaya'    # explicitly char type is not available
print(a, type(a))
```

```python
# Example 5, range, list, tuple, set, frozenset, dict, None
a = range(10)
print("Range: ",a, type(a))
a = (1, 2, 3)
print("Tuple: ",a, type(a))
a = {1, 2, 3}
print("Set: ",a, type(a))
a = frozenset({1, 2, 3})
print("Frozenset: ",a, type(a))
a = {1:'A', 2:'B', 3:'C'}
print("Dictionary: ",a, type(a))
a = None
print("None: ",a, type(a))
```

```python
# Example 6, bytes, bytearray
a = [1,2,3]
print("List: ",a, type(a))

a = bytes('123', 'utf-8')
print(a, type(a))

a = bytearray('123', 'utf-8')
print(a, type(a))

a = bytearray(b"abcd")
print(a, type(a))
```

# Decimal, Binary, Octal, Hexa decimal form

```python
# Example 7, Number systems
a = 100; b = 0B100; c = 0o100; d=0x100; e=0X100
print("Decimal:",a, "Binary:", b, "Octal:",c, "Hex: ", d, e)

# bin(), oct(), hex()
a = bin(77); b = oct(77); c = hex(77)
print("Binary:", a, "Octal:",b, "Hex: ", c)
```

# Type Casting (Type conversion)
* convert one type value to another type
* int(), float(), complex(), bool(), str()

```python
# Example 8, type cast
a = int(100); b = int(10.55); c = int(True); d = int(False) # c = int(5 + 4j); d = int(0B100); 
print(a, b, c, d)

print()
a = float(100); b = float(10.55); c = float(False); # c = float(5 + 4j); d = float(0B100); 
print(a, b, c)

print()
a = complex(100); b = complex(10.55); c = complex(False); d = complex(5, 2) 
print(a, b, c, d)
```

```python
# Example 9, Type cast
a = bool(100); b = bool(10.55); c = bool(False); d = bool(0); e = bool(-5) 
e = bool("True"); f = bool("False")
print(a, b, c, d, e, f)

print()
a = str(100); b = str(10.55); c = str(False); d = str(0); e = str(-5 + 2j); e = str("True")
print(a, b, c, d, e, f)
```

# Mutable vs Immutable
* All Fundamental Data types are immutable.
* If a data type is immutabble then we cannot changes the content and if we try to change the content then with those changes a new object will be created. 
* **Immutable**: int, float, complex, bool, str, bytes, range, tuple, frozenset, None
* **Mutable**: bytearray, list, set, dic


# When to use which variable?
* int: for whole or integral numbers
* float: for decimal or floating point numbers
* complex: for complex numbers
* bool: for logical values (True, False)
* str: for set of Characters


* range: For range of values (imutable)
* List: for ordered collection of values (mutable)
* Tuple: for ordered collection of values (imutable)
* set: for un-ordered collection of values (mutable)
* frozenset: for un-ordered collection of values (immutable)
* dict: for group of key value pairs (mutable)
