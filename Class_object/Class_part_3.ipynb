#### 1. Operator Overloading
* We can use the same operator for multiple purposes, e.g., + is for add and string concatenation.
* For every operator *magic methods* are available. To overload any operator we have to override  that Method in our class. 

```python
#Example 1, operator overloading
class processNumber:
    def __init__(self, num):
        self.number = num

    def __add__(self, other):
        return self.number + other.number
    
n1 = processNumber(10)
n2 = processNumber(3)

print('Output is:',n1 + n2)
```

```python
#Example 2, operator overloading multiple times
class processNumber:
    def __init__(self, num):
        self.number = num
        
    def __add__(self, other):
        return self.number + other.number
    def __sub__(self, other):
        return self.number - other.number
    def __mul__(self, other):
        return self.number * other.number
    def __le__(self,other):
        return self.number <= other.number
    def __eq__(self,other):
        return self.number == other.number
    def __ne__(self,other):
        return self.number != other.number

n1 = processNumber(10)
n2 = processNumber(3)

print('Output is:',n1/n2) 
```

### 2. Method Overloading
* methods having same name but different type of arguments then those methods are said to be overloaded methods.

```python
#Example 3, In python, method overloading is not possible
class myClass:
    def method1(self):
        print('method with no argument')
        
    def method1(self, a):
        print('A method with single argument')
        print("Number is: ",a)
        
    def method1(self, a, b):
        print('A method with two arguments.')
        print("Sum is: ",a+b)

obj = myClass()
#obj.method1()
obj.method1(3)
#obj.method1(2, 3) 
```

```python
# Example 4, How to implement method overloading? 
class myClass:
    def sum(self, *num):
        print("The num is: ",num)
        self.total = 0
    
        for val in num:
            self.total += val
        print('Sum is:', self.total)
        
obj = myClass()

obj.sum(1, 2)
obj.sum(3, 5, 7)
obj.sum(73)
obj.sum() 
```

### 3. Constructor Overloading

* Constructor overloading is not possible in Python.
* If multiple constructors are defined then only the last constructor will be executed.

```python
#Example 5, In python, constructor overloading is not possible
class myClass:
    def __init__(self):
        print('method with no argument')
        
    def __init__(self, a):
        print('A method with single argument')
        print("Number is: ",a)
        
    def __init__(self, a, b):
        print('A method with two arguments.')
        print("Sum is: ",a+b)

obj = myClass(2, 3)
#obj = myClass(2)
```

```python
# Example 6, How to implement constructor overloading?
class myClass:
    def __init__(self, *num):
        print(num)

obj = myClass(1, 2, 3)       
obj = myClass(1,2)
obj = myClass(1)
obj = myClass()
```

## B. Overriding

If a child class is not satisfied with parent class implementation (e.g., method), then child class can be redefine the method in the child class based on its requirement. 

### 1. Method overriding

Overriding concept applicable for both methods and constructors.

```python
# Example 7, method overriding
class vehicle:
    def mileage(self):
        print('Hi there! you are accessing parent class')

class car(vehicle):
     def __init__(self):
        pass
     def mileage(self):
        print('You are accessing child class')
        super().mileage()
        
obj1 = vehicle()
obj1.mileage()

obj2 = car()
obj2.mileage()
```

### 2. Constructor overriding

```python
# Example 8, constructor overriding
class vehicle:
    def __init__(self):
        print('Hi there! you are accessing parent class')

class car(vehicle):
    def __init__(self):
        print('You are accessing child class')
       
obj1 = vehicle()
obj2 = car()
```

```python
# Example 9, constructor overriding application using super
class Person:
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name

class studentDetails (Person):
    def __init__(self, roll, name, age, addr):
        super().__init__(roll, name)
        self.age = age; self.addr = addr
    def details(self):
        print("Student details:", [self.roll, self.name, self.age, self.addr])

e = studentDetails(1,"Ashok", 19, "Mumbai")
e.details()
```

### C.  Duck Typing Philosophy
* Based on provided value at runtime the type of the variable is considered automatically. Hence Python is called "Dynamically Typed Programming (DTP)". 

```python
# Example 10, DTP in class and object
class add:
    def __init__(self,a, b):
        self.a = a; self.b = b      
    def cal(self):
        print(self.a + self.b)
class sub:
    def __init__(self,a, b):
        self.a = a; self.b = b      
    def cal(self):
        print(self.a - self.b)        
class mul:
    def __init__(self,a, b):
        self.a = a; self.b = b      
    def cal(self):
        print(self.a * self.b)
        
def myMethod(obj):
    obj.cal()
    
CL = [add(2, 3), sub(10, 7), mul(11,7)]
for obj in CL:
    myMethod(obj) 
```
