# A.  Abstract Class & Method

* An **abstract class** is a blueprint for other classes. It allows us to define a set of methods that should be constructed within any sub-classes built from the abstract class. 

* Partially implementation classes are called **abstract classes**. In other words, a class that contains one or more abstract methods is called an **abstract class**. 

* Every **abstract class** in Python should be derived from ABC (Abstract Base Classes) class which is present in abc module.

* If a class contains atleast one **abstract method** and if we are extending ABC class then instantiation is not possible i.e., **abstract class** with abstract method instantiation is not possible.

* Note that, parent class abstract methods should be *implemented* in the child classes. Otherwise, we cannot instantiate child class.If we are not creating *child class object*, then we won't get any error.


* An **abstract method** is a method that has a declaration but does not have an implementation.

* Sometimes we don't know about the implementation of any method, but still we need to declare a method, then such type of methods can be declared as **abstract methods**. i.e., it has only declaration part but not implementation part.

* Note that, child classes are responsible to provide implemention for parent class abstract methods. In python we can declare abstract method by using @abstractmethod decorator.

### (A.1) Abstract method

**Syntax of abstract method**:
>from abc import * 

>@abstractmethod \
>def m1(self): 
>> pass

```python
# Example 1
from abc import *
class student:
    
    @abstractmethod
    def marks(self):
        pass 
```

### (A.2) Abstract class

```python
# Example 2 
from abc import *
class myClass:
    pass

obj = myClass() 
print(myClass)
print(obj)
```

```python
# Example 3, Every abstract class should be derived from ABC class
from abc import *

class myClass(ABC):
    pass

obj = myClass() 
print(obj)
```

* If a class contains atleast one **abstract method** and if we are extending ABC class then instantiation (object creation) is not possible i.e., **abstract class** with abstract method instantiation is not possible (Example 4).

```python
# Example 4, Abstract class with abstract method instantiation is not possible 
from abc import *

class myClass(ABC):
    
    @abstractmethod
    def method1(self):
        pass

#obj = myClass() # TypeError: Can't instantiate abstract class Test with abstract methods m1
```

```python
# Example 5, noes: number of edges
#from abc import ABC, abstractmethod
from abc import *
class Polygon(ABC):
    
    @abstractmethod
    def noes(self):
        pass

class Triangle(Polygon): #overriding abstract method
    def noes(self):
        print("It contains 3 edges")

class Pentagon(Polygon):
    pass
    def noes(self):
        print("It contains 5 edges")

obj1 = Triangle(); obj1.noes()
obj2 = Pentagon(); obj2.noes()
```

```python
# Example 6 
from abc import *
 
class myClass(ABC):      
    def myMethod(self):
        pass
 
class myChildClass(myClass):
    def myMethod(self):
        print("abstarct class and method are going on!")
        
obj = myChildClass() 
print("Check is sub-class: ",issubclass(myClass, ABC))
print("Check is sub-class: ",issubclass(myChildClass, myClass))
print( "Check is instance: ",isinstance(obj, myClass))
print( "Check is instance: ",isinstance(obj, myChildClass))
print( "Check is instance: ",isinstance(obj, ABC))
```

Note that, **Concrete classes** contain only concrete methods, but **abstract classes** may contain both concrete methods and abstract methods.(Example 7). 

```python
# Example 7
from abc import ABC, abstractmethod 
class myClass(ABC):                          # abstract base class
    def method1(self):                       # concrete methods
        print("1. Love, tolerance, and selflessness should be practiced.")
        
    @abstractmethod 
    def absMethod(self):                     # abstract methods
        pass
 
class mySubClass(myClass):
    def method1(self):
        super().method1()
        print("2. You are only entitled to the action, never to its fruits.")      
    def absMethod(self):
        print("3. Good work never wasted, always rewarded by the God.")

obj = mySubClass()
obj.method1()
obj.absMethod()
```

### (B.) Interfaces 
* If an abstract class contains only abstract methods such type of abstract class is considered as **interface**.

### Concreate class, Abstract Class, Inteface
* **Concreate class**: completely Implementation and ready to provide service.
* **Abstract class**: partially implemented class i.e., implementation is done partially but not completely.
* **Inteface**: If we don't know anything about implementation, we just mentioned required specification then we should go for interface.

**Interface**: An interface is a blueprint for designing classes. Like classes, interfaces define methods, but these methods are abstract. An abstract method is one that the interface simply defines. It doesnt implement the methods. 

```python
# Example 8, Interface
from abc import *
class myClass(ABC):
    @abstractmethod
    def method1(self): pass
    
    @abstractmethod
    def method2(self): pass
    
    @abstractmethod
    def method3(self): pass

class subClass(myClass):
    def method1(self):
        print('Eena meena deeka')
    
    def method2(self):
        print('Why kattappa killed bahubali?')

class subSubClass(subClass):
    def method3(self):
        print("Success Ke Peeche Mat Bhaago, Excellence Ka Peecha Karo!!")
        
obj = subSubClass()
obj.method1(); obj.method2(); obj.method3() 
```

### (C.) Public, Protected and Private Attributes

**Public attributes**: By default every attribute is public. We can access from anywhere either within the class or from
outside of the class.

**Protected attributes**: Such attrbutes can be accessed within the class anywhere but from outside of the class only in child classes. We can specify an attribute as protected by prefexing with _ symbol.

**Private attributes**: Such attrbutes can be accessed within the class anywhere but from outside of the class only in child classes. We can specify an attribute as protected by prefexing with \_\_ (double underscore) symbol. Note that, we can't access private variables directly from outside of the class. But we can access it e.g., objName._className__variableName

```python
# Example 9, Public, Protected and Private Attributes
class myClass:
    v1 = 3     # Public attribute
    _v2 = 5    # Protected attributes
    __v3 = 7   # Private attribute
    
    def method1(self):
        print("Public variable: ", myClass.v1)
        print("Protected variable: ", myClass._v2)
        print("Private variable: ", myClass.__v3)

obj = myClass()
obj.method1()

print("Outside the class, public variable: ",myClass.v1)
print("Outside the class, protected variable: ", myClass._v2)
#print("Outside the class,private variable: ", myClass.__v3)
print("Outside the class,private variable: ", obj._myClass__v3)

print("Outside the class, using object, public: ",obj.v1)
print("Outside the class, using object, protedted: ", obj._v2)
#print("Outside the class,private variable using object: ", obj.__v3)
```

### (D) \_\_str\_\_() , and repr() functions
* \_\_str()\_\_ method returns a string to representation object and we use it to increase readability. 

```python
# Example 10, use of __str()__ method
class EMP:
    def __init__(self,eno,name):
        self.empno = eno
        self.ename = name       

    def __str__(self):
        return 'It is an object, Employ number :{} and Name:{}'.format(self.empno, self.ename,)

e1 = EMP(1, 'Nikita')
e2 = EMP(2, 'Srikanth')
print(e1); print(e2) 
#print(repr(e1))
```

The **repr()** function is used to return string containing a printable representation of object. Note that, result string can be converted into original object by using **eval() function**, that is not possible in **str() function**.

```python
# Example 11, use of repr() function
import datetime

today = datetime.datetime.now()
print("Output of the object: ",today)
pDate = repr(today) #converting datetime object to str
print(pDate)

dte = eval(pDate)   #converting str object to datetime
print("Date and time is: ",dte) 
```
