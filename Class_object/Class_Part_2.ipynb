## A. Using members of one class inside another class

### 1. Composition (Has-A relationship)
* **Has-A Relationship**: By using class name or object, we can access members of one class inside another class. A class Composite can contain an object of another class Component. 

* It enables creating complex types by combining objects of other types. 

* Advantage is code reusability.

```python
# Example 1, Has A relationship
class myClass1:
    var1 = 3  
    def __init__(self):
        self.var2 = 5
    def method1(self):
        print('I am in class 1')

class myClass2:
    def __init__(self):
        self.var3 = myClass1() # Create object of myClass
    def method2(self):
        print('I am in class 2')
        print("Display var1: ",self.var3.var1)
        print("Display var2: ",self.var3.var2)
        self.var3.method1() # Call method1 of myClass1
        
c = myClass2()
c.method2()
```

```python
#Example 2, Has A relationship
class class1:
    var1 = 3
    def __init__(self):
        self.var2 = 5
    def myMethod1(self):
        print("myMethod1 of class 1")

class class2:
    var3 = 7
    def __init__(self):
        self.var4 = 9
    def myMethod2(self):
        print("myMethod2 of class 2")
    def myMethod3(self):
        x1 = class1()
        print("Value of var 1 and 2: ",x1.var1, x1.var2)
        x1.myMethod1()
        print("Value of var 3 & 4: ", class2.var3, self.var4)
        self.myMethod2()
        print("myMethod3 of class 2")
        
y1 = class2() 
y1.myMethod3()
```

### 2. Inheritance (IS-A relationship)
**Inheritance** is a technique that allows us to inherit all the attributes and methods from another class. The class from which the the attributes and methods are utilized is called the *parent class (base class)*. The class which uses all the attributes and methods from another class is called as *child class (derived class)*.
>class class1:
>>    #methods and variables

>class class2(class1):
>>    #methods and variables

* Methods of parent class (class1) automatically available to the child class (class2)
* **Advantage of inheritance** is code reusability and we can extend existing functionality with some more extra functionality.

* Classes that inherit from another are called derived classes, subclasses, or subtypes and classes from which other classes are derived are called base classes or super classes.

* vehicle class --- car class, car is a vehicle, car inherits properties from the class vehicle.

```python
# Example 3, Inheritance
class class1:
    var1 = 7
    def myMethod1(self):
        print("Parent class (class1) in myMethod1")

class class2(class1):
    var2 = 9
    def myMethod2(self):
        print("Child class (class 2) in myMethod2")

objC2 = class2()
objC2.myMethod1()
objC2.myMethod2()

print("Value of var1: ",objC2.var1)
print("Value of var2: ",objC2.var2)
```

### The super() Method
It is a built-in method which is useful to call the super class constructors, variables and methods from the child class.

```python
# Example 4, example of super() method
class Person:
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
    def studentData(self):
        print('Education is the passport to the future...')

class studentDetails (Person):
    def __init__(self, roll, name, age, addr):
        super().__init__(roll, name)
        self.age = age; self.addr = addr
    def justDisplay(self):
        print("Teachers can open the door, but you must enter it yourself.")
    def details(self):
        print("Student details:", [self.roll, self.name, self.age, self.addr])

e = studentDetails(1,"Ashok", 19, "Mumbai")
e.studentData()
e.justDisplay()
e.details()
```

### Types of Inheritance: 
* (i) Single Inheritance
* (ii) Multi Level Inheritance
* (iii) Hierarchical Inheritance
* (iv) Multiple Inheritance
* (v) Hybrid Inheritance

 #### (i) Single inheritance
 * Inheriting the properties from one class to another class

```python
# Example 5 of Single Inheritance
class class1:
    def method1(self):
        print("I am in parent method")
class class2(class1):
    def method2(self):
        print("I am in child method")

obj1 = class2()
obj1.method1()
obj1.method2() 
```

#### (ii) Multi level inheritance
* Inheriting the properties from multiple classes to single class one after another.

```python
# Example 6 of Multi Level Inheritance
class class1:
    def method1(self):
        print("I am in grandparent method.")

class class2(class1):
    def method2(self):
        print("I am in parent method.")
        
class class3(class2):
    def method3(self):
        print("I am in child method.")

obj1 = class3()
obj1.method1(); obj1.method2(); obj1.method3() 

obj2 = class2(); obj2.method1(); obj2.method2() 
# obj2.method3() # AttributeError: 'class2' object has no attribute 'method3'
```

#### (iii) Hierarchical Inheritance

```python
# Example 7 of Hierarchical Inheritance
class class1:
    def method1(self):
        print("I am in parent method.")

class class2(class1):
    def method2(self):
        print("I am in child1 method.")
        
class class3(class1):
    def method3(self):
        print("I am in child2 method.")

obj1 = class2(); obj2 = class3()

obj1.method1(); obj1.method2(); obj2.method1(); obj2.method3()

# obj1.method3()
# obj2.method2()
```

##### (iv) Multiple Inheritance

* Method Resolution Order(MRO): Way a programming language resolves a method or attribute. 
* In python, MRO defines the order in which the base classes are searched when executing a method.

```python
# Example 8, Multiple Inheritance
class class1:
    def method1(self):
        print("I am in father method.")

class class2:
    def method1(self):
        print("I am in mother method.")
        
class class3(class2, class1):   # Order of parent class is important
    def method3(self):
        print("I am in child method.")

obj1 = class3()
obj1.method1(); obj1.method3()#; obj1.method2()
#obj1.method2()
```

#### (v) Hybrid Inheritance

```python
#Example 9, Hybrid Inheritance
class class1:
    def method1(self):
        print("I am in father method.")

class class2:
    def method2(self):
        print("I am in mother method.")
        
class class3(class1, class2):
    def method3(self):
        print("I am in child method.")
        
class class4(class3):
    def method4(self):
        print("I am in grand child method.")

obj1 = class4()
obj1.method1(); obj1.method2(); obj1.method3(); obj1.method4()
```

```python
#Example 9, Hybrid Inheritance
class class0:
    def method0(self):
        print("I am in grand parent method.")
class class1(class0):
    def method1(self):
        print("I am in father method.")

class class2(class0):
    def method2(self):
        print("I am in mother method.")
        
class class3(class1, class2):
    def method3(self):
        print("I am in child method.")
        
class class4(class3):
    def method4(self):
        print("I am in grand child method.")

obj1 = class4(); obj1.method0(); obj1.method1(); obj1.method2(); obj1.method3(); obj1.method4()
```

By following the method resolution order, the lookup order as follows. Class4 -> Class3 -> Class1 -> Class2 ->Class0. Python follows depth-first order to resolve the methods and attributes. 

#### Aggregation 
* Aggregation is a week form of composition. If you delete the container object, the contents objects can live without container object.

* The lifetime of the owned object does not depend on the lifetime of the owner.

* Aggregation works the best, when there is a condition where the classes might not have a parent-child relationship and rather have a peer relation.

* Aggregation has a unidirectional association and both the associate class objects can survive individually.

```python
# Example 10, Aggregation
class Class1:
    def __init__(self, var1):
        self.var1 = var1
    def myMethod1(self):
        print("I am in class 1, and avr1 is:", self.var1)

class Class2:
    def __init__(self, var1, var2):
        self.var1 = var1
        self.var2 = var2
    def myMethod2(self):
        self.var1.myMethod1()
        print("var 2 is: ", self.var2, self.var1.var1)
  
objC1 = Class1(3)

objC2 = Class2(objC1, 4)
objC2.myMethod2()

objC1.myMethod1()
del objC1
objC2.myMethod2()
```
