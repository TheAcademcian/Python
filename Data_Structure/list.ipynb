# 1. List: Introduction

* Insertion order of data is preserved
* Duplicate data and heterogeneous data are allowed
* List size can be changed
* Supports both positive and negative indexes
* Mutable

```python
# Example 1, Create a list
myList = []
print("List is: ", myList)
print("Type is: ",type(myList)) 
```

```python
# Example 2, more list creation examples
myList = [1, 2, 3, 4]
print("List is: ", myList)

myList = [1, 2, 3, 4, 3, 4] # Duplicate values are allowed
print("List is: ", myList)

myList = ['Amar', 'Akhbar', 'Anthony'] 
print("List is: ", myList)

myList = ['Ganga', 'Jamuna', 'Kaveri', 14, 3.43, 7.007, "I have a dream"] 
print("List is: ", myList)

myList = [True, False, False, True] 
print("List is: ", myList)
```

```python
# Example 3
myList = eval(input("Enter List:")) # Enter data with [1, 3, 4]
print("The list is: ",myList)
print("Type is: ",type(myList)) 
```

```python
# Example 4, splitting a string into words
myString = "Friendship is the wine of life."
myList = myString.split()
print(myList)
```

# List Comprehensions
* Create a list in a compact way
* newlist = [expression for item in iterable if condition == True]

```python
# Example 5,List comprehensions
myList = [x for x in range(-3, 4)]
print("The content of the list is:",myList)
```

```python
# Example 6, List comprehensions
myList = [x**2 for x in range(-2,3)]
print("1. List: ",myList)

myList1 = [x for x in myList if x%2 == 0] 
print("2. List: ",myList1)
```

```python
# Example 7, List comprehensions
colors = ["Violet", "Indigo", "Blue", "Green", "Yellow", "Orange", "Red"]
myList = [col[0] for col in colors]
print("List is: ", myList)
```

```python
# Example 8 Intersection using list comprehensions
List1 = ["Blue", "Green", "Yellow",]
List2 = ["Green", "Yellow", "Orange"]

List3 = [ x for x in List1 if x in List2]
print(List3) 

# Difference using list comprehensions
List1 = ["Blue", "Green", "Yellow"]
List2 = ["Green", "Yellow", "Orange"]
List3 = [ x for x in List1 if x not in List2]
print(List3) 
```

```python
# Example 9 Split a sentense into words and print word length of each splitted words
sentense = "I Have a Dream"
myWords = sentense.split()
print("Words list is:", myWords)

myList = [[w.lower(),len(w)] for w in myWords]
print("Final list is: ", myList) 
```

```python
# Example 10 Example remove a patricular element
place = ["daringbadi", "darjeeling", "kannur", "shimla"]

myList = [x for x in place if x != "shimla"]
print("Final list is: ", myList) 
```

```python
# Example 11 Return only absolute value of -ve values 
myList = [-3, -2, -1, 0, 1, 2, 3]
newList = [abs(x) for x in myList if x < 0]
print("Final list is: ", newList) 
```

# 3. List functions

### len(listName), listName.count(data), listName.index(data)

```python
# Example 20, use of len, count and index

myList = ["Shirdi", "Bhimashankar", "Kolhapur", "Ganpatipule", "Kolhapur"]

print("Size of the list is: ",len(myList))
print("Count : ", myList.count("Kolhapur"))
print("Index : ", myList.index("Ganpatipule"))
```

```python
# Example 20, sorting
myList = [333, 1, 4444, 22, 55555]

myList.sort()
print("The sorted list is: ", myList)

myList.sort(reverse = True)
print("The sorted list is: ", myList)
```

```python
# Example 24, removing a data with if and else
List = ["Soccer", "Tennis", "Golf"]
print("List before deletion: ",List)

x = input("Enter element name to delete: ")

if x in List:
    List.remove(x)
    print("List after deletion: ",List)
else:
    print("Element is not present in the list!") 
```

### listName.pop()

# 5. Aliasing & Cloning 

```python
# Example 27 Aliasing (Renaming)
list1 = ["CV Raman", "Homi J. Bhabha", "Visvesvaraya", "Venkatraman"]
list2 = list1
print(id(list1), list1)
print(id(list2), list2)

list1.append("Abdul Kalam")

print(list1)
print(list2)
```

```python
# Example 28 Cloning
list1 = ["CV Raman", "Homi J. Bhabha", "Visvesvaraya", "Venkatraman"]
list2 = list1.copy()
print(id(list1), list1)
print(id(list2), list2)

list1[3] = "Abdul Kalam" 

print()
print(list1)
print(list2)
```

# 7. List comparisons
* ==, !=, >, >=, <, <=

```python
# Example 32
list1 = ["Red","Green","Blue"]
list2 = ["Red","Blue","Green"]
list3 = ["Red","Green","Blue"]
list4 = ["Red","Green"]
list5 = ["Green","Blue","Red"]

print("list1 == list2: ",list1 == list2) 
print("list1 == list3: ",list1 == list3) 
print("list1 != list3: ",list1 != list3) 
print("list1 == list4: ",list1 == list4) 
print("list1 == list5: ",list1 == list5) 
```

```python
# Example 33 Compare only the first element
list1 = [3, 5, 7]
list2 = [3, 5, 7]
list3 = [5, 3, 7]

print("1. list1 > list2: ", list1 > list2) 
print("2. list1 > list2: ", list1 >= list2) 

print("3. list1 > list3: ", list1 > list3) 
print("4. list1 > list3: ", list1 >= list3) 

print("5. list2 > list3: ", list2 > list3) 
print("6. list2 > list3: ", list2 >= list3) 
```
