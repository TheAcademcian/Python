# String formatting
> using {} and format() 
* The curly braces are used as a placeholder in the string, and the format() function argument is utilised to replace them.

```python
# Example 23
a, b, c =  10, 20, 30
print("The content is {}, {}, {}. Thank You!".format(a,b,c))
```

```python
# Example 24, Numbers Formatting, format() for integer numbers
print("The int no. is: {}".format(777))
print("The int no. is: {:d}".format(777))
print("The int no. is: {:d}".format(-777))
print("The int no. is: {:6d}".format(777))
print("The int no. is: {:06d}".format(777))

print("\nThe int no. is: {:+d}".format(777)) # for signed numbers
print("The int no. is: {:+d}".format(-777))
```

```python
# Example 25, format() for floating numbers
print("1.The float no. is: {}".format(1357.9))
print("2.The float no. is: {:.2f}".format(1357.91345))
print("3.The float no. is: {:6.3f}".format(1357.9))
print("4.The float no. is:  {:6.3f}".format(1357.900))
print("5.The float no. is:  {:06.3f}".format(13579.999))
print("6.The float no. is:  {:6.3f}".format(12345678.9))
#  default alignment is Right Alignment

print("\n7.The float no. is: {:+.2f}".format(1357.9))
print("8.The float no. is: {:+6.3f}".format(-1357.9))
```

```python
# Example 26, format() for binary, octal, decimal numbers
print("Binary value of 777 is:{0:b}".format(777))
print("Octal value of 777 is:{0:o}".format(777))
print("Hex. value of 777 is :{0:x}".format(756))
print("Hex. value of 777 is :{0:X}".format(756))
# print("Binary value of 777 is:{0:b}".format(777.77)) # ValueError
# print("Octal value of 777 is:{0:o}".format(777.12))
print("Binary value of 777 is:{0:+b}".format(-777))
print("Binary value of 777 is:{0:+b}".format(777))
```

```python
# Example 27, string alignment
print("1.", "{:5d}".format(777))
print("2.", "{:<5d}".format(777))  # Left Alignment
print("3.", "{:<05d}".format(777)) # Left Alignment
print("4.", "{:>5d}".format(777))  # Right alignment
print("5.", "{:>05d}".format(777)) # Right alignment
print("6.", "{:^5d}".format(777))  # Center alignment
print("7.", "{:^05d}".format(777))  # Center alignmen
print("8.", "{:=5d}".format(-777)) # Put sign symbol to the left most position

print("9.", "{:^ 10.3f}".format(77.1234))
print("10.", "{:= 6.3f}".format(-77.1357)) 
```

```python
# Example 28, format() for strings
print("1.", "{:7}".format("MIT"))
print("2.", "{:>7}".format("MIT")) # Right alignment
print("3.", "{:<7}".format("MIT")) # Left alignment
print("4.", "{:^7}".format("MIT")) # Center alignment
print("5.", "{:*^7}".format("MIT"))
print("6.", "{:@^7}".format("MIT"))

#print("2.", "{:7d}".format("MIT")) # ValueError
# Default alignment for no. is right, but default alignment for strings is left
```

```python
# Example 29, Truncating Strings with format() method
print("1.", "{:.4}".format("ABCDE"))
print("2.", "{:10}".format("ABCDE"))
print("3.", "{:>10.4}".format("ABCDE"))
print("4.", "{:^10.3}".format("ABCDE"))
print("5.", "{:*^10.2}".format("ABCDE")) 
```

# Few interesting string programs

```python
# Example 36, String reversing
myStr = input("Enter your string: ")

revMyString = myStr[::-1]
print("Reversed string is:",revMyString)
```

```python
# Example 37, String reversing (another way)
myStr = input("Enter your string: ")

i = len(myStr)-1

revMyString = ''  # initialization

while i>=0:
    revMyString += myStr[i]
    i -=1

print("Reversed string is:",revMyString)
```

```python
# Example 38, String reversing (another way)
myStr = input("Enter your string: ")

wordList = myStr.split()
print(wordList)

revWordList = []

i=len(wordList)-1
while i>=0:
    revWordList.append(wordList[i])
    i -=1
    
print("Reversed string in list form is:", revWordList)

revWordList = ' '.join(revWordList)
print("Reversed string is:", revWordList)
```

```python
# Example 39, myStr: ABC123XYZ567, finalStr: ABCXYZ123567
myStr = input("Enter Some String:")

myStr1 = myStr2 = ''

for char in myStr:
    if char.isalpha():
        myStr1 += char
    else:
        myStr2 += char
        
print("myStr1: ", myStr1)
print("myStr2: ", myStr2)

finalStr =' '
for char in myStr1:
    finalStr += char
    
for char in myStr2:
    finalStr += char
    
print("The final output is: ", finalStr) 
```

```python
# Example 40, myStr: A7B2C3, FinalResult: AAAAAAA BB CCC
myStr = input("Enter string:")

FinalResult = ''

for char in myStr:
    if char.isalpha():
        FinalResult += char
        lastChar = char 
    else:
        FinalResult += lastChar * (int(char)-1)
        FinalResult +=' '

print("The final result is: ", FinalResult) 
```

```python
# Example 41, Remove duplicate characters from a string, e.g., myStr: AABBBCCS and fResult: ABCS
myStr = input("Enter string: ")

CharList = []

for x in myStr:
    if x not in CharList:
        CharList.append(x)

print("In list",CharList)
fResult = ''.join(CharList)      
print("Final result:",fResult) 
```

```python
# Example 42, myStr: academician, Final result: a:3, c:2, d:1, e:1, m:1, i:2, n:1, 
myStr = input("Enter string: ")

dictionary = {}

for key in myStr:
    if key in dictionary.keys():
        dictionary[key] += 1
    else:
        dictionary[key] = 1

print("\nFinal result: ", end = '')
for key,val in dictionary.items():
    print("{}:{}, ".format(key, val), end = '') 
```
