# Dictionary

* In the dictionary, we store data as **key - value** pairs, dic = {key : value}, 
* e.g., dic = {key1 : value1, key2 : value2, ..., keyN : valueN}
* Duplicate keys are not allowed, but duplicate values are allowed.
* Both keys and values can contain heterogeneous values.
* The order of insertion is not preserved.
* Indexing and slicing are not allowed.

# Basic functions of dictionary
dict(), clear(), len(), get(), pop(), popitem(), keys(), values(), items(), copy(), setdefault(), update()

```python
# Example 9, len()
myDict = {1:'Oatmeal' , 2: 'Watermelon', 3:'Sweet Potatoes', 4:'Water'}
print("Size of the dictionary is: ", len(myDict))
```

```python
# Example 10, get()
myDict = {1:'Oatmeal' , 2: 'Watermelon', 3:'Sweet Potatoes'}

print("1.The content (value) of the said key is", myDict.get(1))
print("2.The content (value) of the said key is", myDict.get(5, "Food"))
# Set default value is Player
print("3.Dinctionary:", myDict)
```

```python
# Example 11, pop() and popitem()
myDict = {1:'Oatmeal' , 2: 'Watermelon', 3:'Sweet Potatoes', 4:'Water'}
print("1.The content (value) of the said key is",myDict)

print("2.The content (value) of the said key is",myDict.pop(3)) 
# for removing any particular item
print("3.The content (value) of the said key is", myDict)

print("\n4.The content (value) of the said key is",myDict.popitem()) # for removing any item
print("5.The content (value) of the said key is", myDict)

print("\n6.The content (value) of the said key is",myDict.popitem()) # for removing any item
print("7.The content (value) of the said key is", myDict)
```

```python
# Example 12 keys() and values()
myDict = {"Food1":'Mangoes', "Food2": 'Greek Yogurt', "Food3":'Leafy Greens'}
print("1. The content of the dictionary", myDict)
print("2. The content (value) of the said key is", myDict.keys())

for i in myDict.keys():
    print("Key: ", i)
        
print("\n3. The content (value) of the key is\n", myDict.values())

for i in myDict.values():
    print("Value: ", i)
```

```python
# Example 15, Merging two dictionaries
myDict1 = {1:'Larry Bird' , 2: 'Bill Russell', 3:'Oscar Robertson'}
myDict2 = {4:'Magic Johnson' , 5: 'Michael Jordan'}

myDict1.update(myDict2)
myDict2.update(myDict1)

print("\nThe content of the dictionary 1 is:\n", myDict1)
print("\nThe content of the dictionary 2 is:\n", myDict2)
```

```python
# Example 17, Search number of occurrences of each letter present in the given string 
word=input("Enter any word: ")
myDict={}

for c in word:
    myDict[c] = myDict.get(c,0) + 1
    #print("The dictionary is:",myDict)

print("The dictionary is:",myDict)
for m, n in myDict.items():
    print(m, "Appeared ", n, " times") 
```

```python
# Example 18, Count number of vowels present in a given screen
myWord = "Academician"
vowels = {'a','e','i','o','u'}
dictionary = {}
for key in myWord:
    #print("Key:", key)
    if key in vowels:
        dictionary[key] = dictionary.get(key,0)+1

for k,v in sorted(dictionary.items()):
    print(k,"occurred ",v," times")
```

```python
# Example 19, Dictionary Comprehension
Dict1 = {x:x**x for x in range(1,6)}
print(Dict1)

Dict2 = {x:3**x for x in range(-2,3)}
print(Dict2) 
```
