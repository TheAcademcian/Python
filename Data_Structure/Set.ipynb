# Set in Python

## 1. Introduction
* To store  group of unique values 
* Order of the elements is immeterial
* Indexing and slicing not applicable
* Mutable
* All the elements should be distinct, and heterogeneous elements are allowed
* Oparations like union, intersection, difference can be applied

**Syntax**:
> setName = {e1, e2, ..., eN}   # <class 'set'>


```python
# Example 1 Set definition
mySet = {1, 3, 5, 3, 3, 5, 5, 5, 5, 5, -5, -10, 3.55, 'A', "Great"}
print("Display content of the set: ", mySet)
#print("Display content of the set: ", mySet[3:6])
```

**Syntax**:
> set(any sequence)

```python
# Example 2, List to set conversion
list = [1, 3, 5, 2, 4, 6]
mySet = set(list)                            # Conversion list to set
print("1. The content of the set is:", mySet) 

mySet = set(range(5))
print("2. The content of the set is:", mySet)

#mySet = {"Computer", "Civil", "Mechanical", 'Electronics','Bio Technology'}
mySet = {"Computer", "Civil"}

print("3. The content of the set is:", mySet)
print("4. The size of the set is: ", len(mySet))
```

```python
# Example 3, other types set formation, set((e1, e2, ..., en))
mySet = set(("Bandish Bandits", "Panchayat", "Kota Factory",'Man'))
print("The set is:", mySet)

# Set formation with diverse data types
my_set = {5.0, "Three Idiots", (10, 20, 30)}
print("The set is:", my_set)
```

## 2. Basic operations

### (a) add() function
* To add elements into a set

```python
# Example 5, Append an element
mySet = set()
mySet.add('AVATAR')

mySet.add('Inception')
mySet.add('The Godfather')
#mySet.add('No Time To Die','Coco') # Type error

print("The content of the set is: ", mySet) 

# mySet.add("Schindler's List")
# print("The content of the set is: ",mySet) 
```

### (b) update() function
* To add sets of data into another set

```python
# Example 6, Set update
set_1 = {10,20,30} # Set
set_2 = ['The Matrix',' Saving Private Ryan',' Life Is Beautiful'] # List
set_1.update(set_2, range(5)) 
#set_1.update(40)

print("The set is: ",set_1) 
```

### (c) copy() function
* To copy a set into another set, (copy entire elements)

```python
# Example 7, copy content of a set
mySet_1 = {'Messi', 'Ronaldo', 'Neymar'}
mySetCopied = mySet_1.copy()

print("The original set: ", mySet_1)
print("The copied set: ", mySetCopied)
```

### (d) pop() function
* To delete first element

```python
# Example 8, delete element from the begining of a set
mySet = {'Messi', 'Ronaldo', 'Neymar','Zidane'}

print("1.Before pop_1", mySet)
print("The deleted element: ", mySet.pop())
print("1.After pop_1", mySet)

print("The deleted element: ", mySet.pop())
print("1.After pop_2", mySet)

# Another set with same content
mySet = {'Messi', 'Ronaldo', 'Neymar','Zidane'}
print("\n2.Before pop_2", mySet)

print("2.Deleted element: ", mySet.pop())
print("2.After pop_2",mySet)
```

### (e) remove(element)
* To remove any particular element from a set

```python
# Example 9, remove element from a set
mySet = {'Messi', 'Ronaldo', 'Neymar', 'Zidane'}
print("1.The content of the set before is:\n", mySet) 

mySet.remove('Neymar')
print("2.The content of the set after remove is:\n",mySet)

#mySet.remove("Maradona")                               # KeyError
# print("The content of the set is: ", mySet) 
```

```python
mySet = {3, 4, 5, 6}

ele = int(input("Enter the element to delete:"))

if ele in mySet:
    mySet.remove(ele)
    print("The content of the set after remove is:\n",mySet)
else:
    print("The element is not present in the set!")    
```

```python
# Example 10, remove using try and except
mySet = {'Lee','Chen','Lin','Sindhu'}
element = (input("Enter the element to delete: "))

try:
    mySet.remove(element)
    print("The element is removed!")
    print("The content of the set after remove is: ", mySet)
except:
    print("The entered element is not present in the set!")
```

### (f) discard() function 

```python
# Example 11, delete an element using discard function
mySet = {22, 333, 4444, 55555, 666666}
element = int(input("Enter the element to delete: "))
print("The content of the set is: ", mySet) 

mySet.discard(element)
print("The content of the set is: ", mySet) 
```

### (g) clear() function

```python
# Example 12, delete all the elements present in a set
mySet = {'To Kill a Mockingbird', 'The Great Gatsby', 'One Hundred Years of Solitude'}
print("The content of the set before clear:\n", mySet)

mySet.clear()
print("\nThe content of the set after clear is: ", mySet)
```

#  3. Mathematical Operations
  (a). union,
  (b). intersection,
  (c). difference,
  (d). symmetric_difference

### (a) Union()

```python
# Example 13, union
mySet1 = {1, 2, 3, 4}
mySet2 = {10, 20}

print("1.The content of the set is: ", mySet1.union(mySet2)) 

print("2.The content of the set is: ", mySet1|mySet2) 
print("3.The content of the set is: ", mySet1|mySet1) 
```

### (b) Intersection()

```python
# Example 14, intersection
mySet1 = {1, 2, 3, 4}
mySet2 = {3, 4, 5, 6}

print("1.The content of the set after intersection is: ", mySet1.intersection(mySet2)) 

print("2.The content of the set after intersection is: ", mySet1 & mySet2) 
print("3.The content of the intersection set is: ", mySet1 & mySet1) 
```

### (c) difference()

```python
# Example 15, difference
mySet1 = {1, 2, 3, 4}
mySet2 = {3, 4, 5, 6}

print("1.The content of the set after difference is: ", mySet1.difference(mySet2)) 

print("2.The content of the set after difference is: ", mySet1 - mySet2) 
print("3.The content of the difference set is: ", mySet1 - mySet1) 
print("4.The content of the difference set is: ", mySet2 - mySet1) 
```

### (d) symmetric_difference()

```python
# Example 16, Returns elements present in either x or y but not in both
mySet1 = {"Dropbox", "Python", "Microsoft Office PowerPoint"}
mySet2 = {"JIRA", "Python", "Gmail"}

print("1. The difference is: ",mySet1.symmetric_difference(mySet2)) 
print("2. The difference is: ",mySet1^mySet2) 
```

```python
# Example 20, To eliminate duplicate element present in the list (using in and not in)?
mylist = eval(input("Enter List of values: "))

mylist_1 = []
for x in mylist:
    if x not in mylist_1:
        mylist_1.append(x)

print("The content of the set is: ",mylist_1) 
```

# 6. Set comparisons

* intersection_update()	Removes the items in this set that are not present in other, specified set(s)
* isdisjoint()	Returns whether two sets have a intersection or not
* issubset()	Returns whether another set contains this set or not
* issuperset()	Returns whether this set contains another set or not

## (i) intersection_update() 

```python
# Example 22,Removes the items in this set that are not present in other, specified set(s)
setX = {"A Passage to India", "Invisible Man", "Beloved"}
setY = {"Beloved", "Things Fall Apart", "The Color Purple"}

setX.intersection_update(setY)

print("The output is: ", setX)
```

```python
# Example 23
setX = {"badminton", "football", "shooting", "wrestling"}
setY = {"wrestling", "boxing", "tennis", "football"}
setZ = {"squash", "football", "gymnastics", "wrestling"}

setX.intersection_update(setY, setZ)

print("The output is: ", setX)
print("The output is: ", setX & setY & setZ)
```

##  (ii) isdisjoint()

```python
# Example 24, Example of isdisjoint(): 
setX = {"apple", "banana", "cherry"}
setY = {"google", "microsoft", "apple"}
print("The output is: ", setX.isdisjoint(setY))
```

##  (iii) issubset()

```python
# Example 25, issubset()
setX = {"Burj Khalifa", "Shanghai Tower","One World Trade Center"}
setY = {"Shanghai Tower","One World Trade Center"}

print("setX is subset of setY: ", setX.issubset(setY))
```

##  (iv) issuperset()

```python
# Example 26, issuperset()
setX = {"Burj Khalifa", "Shanghai Tower","One World Trade Center"}
setY = {"Shanghai Tower","One World Trade Center"}
print("setX is superset of setY: ", setX.issuperset(setY))
```

### Set comparisons using comparison operators
We can check whether a set is a subset, superset, or equivalent to another set by using the comparison operators, >, =, >=, !=, and == with sets in Python. Depending on the elements in the sets, the boolean true or false is returned.

```python
# Example 27, Set comparisons using comparison operators
set1 = {1 ,2 ,3, 4}; set2 = {1, 3}; set3 = {5, 6, 7}; set4 = {3, 1}
     
print("1.First set is the superset of the second set: ", set1 > set2)  
# True if set1 is the superset of set2 

print("2.First set is the subset of the second set: ", set1 < set2)  
# True if set1 is the subset of set2  

print("3.First set is equivalent to the second set: ",set1 != set2) 
# True if set1 is not equivalent to set2 

print("4.First set is equivalent to the second set: ",set1 == set2) 
# True if set1 is equivalent to set2  

print("5.First set is equivalent to second set: ",set2 == set4) 
# True if set1 is equivalent to set2 
```

# 7. FrozenSets
The frozen sets are the immutable (i.e., the items of the frozen set cannot be changed) sets. Moreover, We cannot change or append the content of the frozen sets by using the functions like add() or remove().

```python
# Example 28, Frozen set
myFSet = frozenset(["Sachin","Dhoni","Kapil", "Murli"])     
print(type(myFSet)) 

print("\nThe content of frozen set is: ")    
for i in myFSet:    
    print(i)    

#myFSet.add(6) 
#gives an error since we cannot change the content of Frozenset after creation    
```
