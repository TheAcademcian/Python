# Tuple

# 1. Tuple: Introduction
* Tuple is immutable
* Order of the data is based on order of insertion
* Allows duplicate data

### (a) How to define a tuple?

```python
# Example 1 tuple data structure
x = 9
y = 9, 2, 11

print("x: ",x)
print("Type of x",type(x))

print("\ny: ",y)
print("Type of y is: ", type(y))

blankT = ()  # Tuple initialization
print("\nType of blankT",type(blankT)) 
```

```python
# Example 2 tuple data structure
x = (7)
print("x: ",x)
print("Type of x",type(x))

y = (7,)
print("\ny: ",y)
print("Type of y is: ", type(y))

z = 7,
print("\nz: ",z)
print("Type of z is: ", type(y))
```

```python
# Example 3, more ways to form tuple
myList = ["Jerusalem", "Wat Rong Khun", "Angkor Wat"]               # List
myTuple = tuple(myList)                                             # List to Tuple
print("1.Content of myTuple 1 is: ", myTuple)

myTuple = tuple(["Singapore", "Hong Kong", "Gibraltar"])
print("2.Content of myTuple 2 is: ", myTuple)

myTuple = tuple(range(3))
print("3.Content of myTuple 3 is: ", myTuple)

myTuple = ("Academician", [3.41, 2.71], (7,77,777), 101)
print("4.Content of myTuple is: ", myTuple)
```

### (b) How to access tuple?
Syntax:
> TupleName[index]

```python
# Example 4, tuple access using index, Index error
myTuple = tuple(["Jerusalem", "Wat Rong Khun", "Angkor Wat"])

index = int(input("Enter the index number: "))
print("The content of th etuple is: ", myTuple[index])
```

```python
# Example 5, tuple access using index and try: except
myTuple = tuple(["Nauru", "Monaco", "Tuvalu"])

index = int(input("Enter the index number: "))

try:
    print("The content of the tuple is: ",myTuple[index])
    
except: 
    print("IndexError: tuple index is out of range!")
```

```python
# Example 6, tuple access using index slicing
myTuple = tuple(["Chocolate","Ice Cream", "French Fries", "Pizza", "Cheese Burger"])

index1 = int(input("Enter the first index: "))
index2 = int(input("Enter the last index: "))

print("The content of the tuple is: ",myTuple[index1:index2])
```

```python
# Example 7, tuple access using index slicing (slice step)
myTuple = tuple(["Chocolate","Ice Cream", "French Fries", "Pizza", "Cheese Burger"])

index = int(input("Enter the index: "))
print("1. The content of the tuple is: ",myTuple[:index]) # Display from the begining to index

print("2. The content of the tuple is: ",myTuple[index:]) # Display from the index to end

print("3. The content of the tuple is: ",myTuple[::index]) # slice step should not be 0
```

```python
# Example 8, tuple access using negative index 
myTuple = tuple(["Chocolate","Ice Cream", "French Fries", "Pizza", "Cheese Burger"])
index = -2

print("1. The content of the tuple is: ",myTuple[index]) 
print("2. The content of the tuple is: ",myTuple[:index]) 
print("3. The content of the tuple is: ",myTuple[index:]) 

print("4. The content of the tuple is: ",myTuple[-5:-1]) 
```

### (c) How to change content of the tuple?

```python
# Example 9, change content of the tuple!!
myTuple = tuple(["Shamosha","Jaleby", "Laddu", "Kachory", "Idli"])

index = int(input("Enter the index: "))
myTuple[index] = "Pizza"      # TypeError: 'tuple' object does not support item assignment 
```

### (d) Check if item is there or not?

```python
# Example 10
myTuple = tuple(["Milk Chocolate","Dark Chocolate", "Cocoa Powder"])

if "Milk Chocolate" in myTuple:
    print("Yes! it is there!")
else:
    print("No! Not there.")
```

### (e) Application of mathematical operators in tuple
* Only \+ and *

```python
# Example 11, Merging two lists
myTuple1 = tuple(["Jennifer Lopez","Jay Leno", "Madonna"])
myTuple2 = tuple(["Angelina Jolie","Chris Brown", "Madonna"])

myTuple  = myTuple1 + myTuple2

print("The content of the tuple is: ", myTuple)
```

```python
# Example 12, Adding an element in a tuple
myTuple = tuple(["Jennifer Lopez","Jay Leno", "Madonna"])
myTuple1 = ("Sunny",)

print("Before: ",id(myTuple))
myTuple += myTuple1 
print("After : ",id(myTuple))

print("The content of the tuple is: ", myTuple)
```

### (f) Convert tuple to list, and lis to tuple


```python
"""Example 13, Other way: convert tuple to list, then append an element, 
then convert back list to tuple"""

myTuple = tuple(["Jennifer Lopez","Jay Leno", "Madonna"])
print("1. My tuple is: ", myTuple)

myList = list(myTuple)
print("2. My List is: ", myList)

myTuple = tuple(myList)
print("3. My tuple is: ", myTuple)
```

```python
# Example 14 use of *
myTuple = tuple(["Bezos", "Ronaldo"])
print("1. The content of the tuple is: ", myTuple)

myTuple = myTuple * 2
print("2. The content of the tuple is:\n", myTuple)
```

# 2. Useful functions in tuple
* len(): len(tupleName)
* count(): tupleName.count(element)
* index(): tupleName.index(element)
* sorted(): sorted(tupleName)
* min(): min(tupleName)
* max(): max(tupleName)

```python
# Example 15, applications of few functions
myTuple = tuple(["Elon Musk", "Jeff Bezos", "Cristiano Ronaldo", "Elon Musk", "Johnson"])

print("1. Apply len(): ", len(myTuple))
print("2. Apply count(): ", myTuple.count("Elon Musk"))
print("3. Index is: ", myTuple.index("Cristiano Ronaldo")) # "Elon Musk"??
print("4. Sorted tuple is: ", sorted(myTuple))

myTuple = tuple([9, 99, 7, 77])
print("\n5. Max value is: ", max(myTuple))
print("6. Min value is: ", min(myTuple))
```

# 3. Packing and Unpacking

```python
# Example 16 Packing
x = "Selena Gomez"; y = "Emma Watson"; z = "Priyanka Chopra"
myTuple = x, y, z

print("After packing the tuple is: ", myTuple) 

# Unpacking
A, B, C = tuple(["Fan Binbing", "Taylor Swift", "Emma Watson"]) # Can we assign more or less variable

print("\nA is: ", A) 
print("B is: ", B) 
print("C is: ", C) 
```

```python
# Example 17 Unpacking using * with variable length
A, *B = tuple(["Fan Binbing", "Taylor Swift", "Emma Watson"]) 
# Can we assign more or less variable

print("\nA is: ", A) 
print("B is: ", B) 
```

# 4. Tuple iteration

```python
# Example 18, Tuple iteration
myTuple = tuple([1,3,5,7])
mul = 1

for e in myTuple:
    print("Tuple: ", e)
    mul *= e
    
print("The mul value is: ", mul)
```

```python
# Example 19, Tuple iteration
myTuple = tuple([1,3,5,7])

for i in range(len(myTuple)):
    print("Tuple: ",myTuple[i])
```

# 5. Delete a tuple

```python
# Example 20, Tuple deletion
myTuple = tuple([1,3,5,7])
print("Tuple is: ", myTuple)

del myTuple
print("Tuple is: ", myTuple) # NameError: name 'myTuple' is not defined
```
